```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

def clean_clinical_data_auto(filepath):
    df = pd.read_excel(filepath)

    #Convert columns to string, clean from commas and unknowns.
    df = df.applymap(lambda x: str(x).replace(",", ".") if pd.notnull(x) else x)
    df.replace(["Not Raised", "Unknown"], np.nan, inplace=True)

    def try_float(x):
        try:
            return float(x)
        except:
            return x
    df = df.applymap(try_float)

    #Change to binary conversion (Yes is 1, No is 0)
    df = df.applymap(lambda x: 1 if str(x).strip().lower() == "yes" else (0 if str(x).strip().lower() == "no" else x))

    for col in df.columns:
        if "Date" in col or "Visitation" in col:
            df[col] = pd.to_datetime(df[col], errors="coerce")

    #Label numeric columns for scaling.
    numeric_cols = df.select_dtypes(include=[float, int]).columns
    
    #Fills NaN with median just for scaling.
    imputer = SimpleImputer(strategy='median')
    scaler = StandardScaler()

    df_scaled = df.copy()
    df_scaled[numeric_cols] = scaler.fit_transform(imputer.fit_transform(df_scaled[numeric_cols]))

    return df

df_clean = clean_clinical_data_auto("Clinical_Chosen.xlsx")
print(df_clean.head())
```
